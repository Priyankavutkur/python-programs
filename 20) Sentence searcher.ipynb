```python
text="I have a cat.I have a mat. Things are going well"
```

```python
def sentence_search(text, search_word):
  if search_word.lower() not in text.lower():
    return " "
  else:
    text=text.split(".")
    for i in text:
      if search_word.upper() in i.upper() or search_word.lower() in i.lower():
        return i+"."
        break
```

```python
sentence_search(text,"MAT")
```
