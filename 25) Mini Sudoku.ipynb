```python
list_1=[[1,2,3],[4,5,6],[7,9,9]]
def is_mini_sudoku(original_list):
  if len(original_list)==0 or len(original_list)==1:
    return False
  else:
    new_string=','.join(str(x) for x in original_list)
    new_string=new_string.replace("[","")
    new_string=new_string.replace("]","")
    new_string=new_string.split(",")
    new_list=[int(x) for x in new_string]
    new_list.sort()
    if len(new_list)!=9 or 0 in new_list:
      return False
    else:
      count=1
      for i in new_list:
        if new_list.count(i)>1:
          count+=1
          break
      if count>1:
        return False
      else:
        return True 
```

```python
is_mini_sudoku(list_1)
```
