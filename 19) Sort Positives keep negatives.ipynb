```python
def reverseString(list1):
  if len(list1)==0:
    return list1
  else:
    index = -1
    for i in range(len(list1)-1, int(len(list1)/2), -1):	
      if list1[i]>=0:
        temp = list1[i]
        while True:
          index += 1
          if list1[index]>=0:
            list1[i] = list1[index]
            list1[index] = temp
            break
    return list1
```

```python
reverseString([])
```
