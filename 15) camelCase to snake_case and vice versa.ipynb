```python
def to_camel_case(string):
  string=string.split("_")
  for i in range(1,len(string)):
    list1=list(string[i])
    list1[0]=list1[0].upper()
    string[i]=''.join(list1)
  return "".join(string)
import re
def to_snake_case(string):
  res_list=re.split("([A-Z][^A-Z]*)", string)
  for i in range(1,len(res_list)):
    if len(res_list[i])==0:
      continue 
    else:
      list1=list(res_list[i])
      list1.insert(0,"_")
      list1[1]=list1[1].lower()
      res_list[i]="".join(list1)
  return "".join(res_list)
```

```python
string="pri_ya_nka"
to_camel_case(string)
```
