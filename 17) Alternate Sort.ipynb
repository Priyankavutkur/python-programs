```python
elements=[1,"A","b",2,"c",3,"a",4,"g"]
def alternate_sort(elements):
  all_list=[]
  string_list=[]
  num_list=[]
  for i in elements:
    if type(i)==str:
      string_list.append(i)
    else:
      num_list.append(i)
  num_list.sort()
  string_list.sort()
  length= lambda a,b: min(a,b)
  length=length(len(string_list),len(num_list))
  for i in range(0,length):
    all_list.append(num_list[0])
    all_list.append(string_list[0])
    num_list.remove(num_list[0])
    string_list.remove(string_list[0])
  if len(string_list)==0 and len(num_list)==0:
    pass
  else:
    all_list+=num_list+string_list
  return all_list
alternate_sort(elements)
```
