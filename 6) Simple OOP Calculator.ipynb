```python
import operator
class Calculator:  
  def add(self,num1,num2):
    return operator.add(num1,num2)
  def sub(self,num1,num2):
    return operator.sub(num1,num2)
  def mul(self,num1,num2):
    return operator.mul(num1,num2)
  def div(self,num1,num2):
    while num2==0:
      num2=int(input("Enter 2nd number grater then 0 : "))
    return operator.truediv(num1,num2)
  def mod(self, num1,num2):
    return operator.mod(num1,num2)
  def xor(self, num1, num2):
    return operator.xor(num1,num2)
```
