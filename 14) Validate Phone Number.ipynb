```python
def is_valid_phone_number(string):
  res = [int(i) for i in string if i.isdigit()]
  if len(res)!=10:
    return False
  elif  "(" and ")" not in string or string[string.index(")")+1]!=" " or "-" not in string:
    return False
  else:
    string=string.split(")")
    for i in string:
      if "(" in i:
        new_string=i.replace("(","")
        if len(new_string)!=3:
          return False
      else:
        new_string=i.split("-")
        new_string[0]=new_string[0].replace(" ","")
        if " " in new_string[0] or " " in new_string[1][0]:
          return False
        else:
          if len(new_string[0])==3 and len(new_string[1])==4:
            return True
          else:
            return False
```

```python
is_valid_phone_number("(123) 4234-7514")
```
