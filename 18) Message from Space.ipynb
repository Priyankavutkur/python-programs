```python
def repeat(word, m, n):
    if(m > len(word)):
        m = len(word)
    repeat_word = word[:m]
    return repeat_word*n
import re
def list_index(text):
  list1=[]
  for i in range(int(len(text))):
    if text[i]=="[" or text[i]=="]":
      list1.append(i)
  return list1
```

```python
def space_message(text):
  index_1,index_2=0,0
  if "[" and "]" not in text:
    return text
  else:
    list1=list_index(text)
    string=text[:list1[0]-1]
    for i in range(0,len(list1),2):
      temp = re.compile("([0-9]+)([a-zA-Z]+)")
      word=text[list1[i]+1:list1[i+1]]
      res = temp.match(word).groups()
      string+=repeat(res[1], len(res[1]), int(res[0]))
      if i<int(len(list1)/2):
        string+=text[list1[i+1]+1:list1[i+2]]
    return string
```

```python
text="pr[5cd]pp[4tt]cc[5p]"
space_message(text)
```
