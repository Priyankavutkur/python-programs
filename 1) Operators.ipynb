```python
import operator

ops = {
    '+' : operator.add,
    '-' : operator.sub,
    '*' : operator.mul,
    '/' : operator.truediv,  # use operator.div for Python 2
    '%' : operator.mod,
    '^' : operator.xor,
}

def Operators(num1, num2,oper):
    op1, op2 = int(num1), int(num2)
    return ops[oper](op1, op2)
```

```python
num1=input("Enter a 1st number: ")
num2=input("Enter 2nd number: ")
oper=input("enter Operators: ")

while num2=="0" and oper=="/":
  num2=input("enter 2nd number grater then 0 :")
Operators(num1,num2,oper)
```
